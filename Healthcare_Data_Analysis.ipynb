```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
print("Libraries loaded!")
```

```python
df = pd.read_csv("/content/healthcare_dataset.csv")
print("Total rows:", len(df))
print(df.head())
```

```python
print(df.columns.tolist())
```

```python
query1_conditions = df.groupby("Medical Condition").agg(
    total_patients=("Medical Condition", "count"),
    avg_billing=("Billing Amount", "mean"),
    avg_stay_days=("Length_of_Stay", "mean")
).reset_index()

print(query1_conditions)
```

```python
query2_admission = df.groupby("Admission Type").agg(
    total_patients=("Admission Type", "count")
).reset_index()

print(query2_admission)
```

```python
df["Age Group"] = pd.cut(df["Age"],
    bins=[0, 20, 40, 60, 100],
    labels=["Under 20", "20-40", "41-60", "Above 60"])

query3_age_group = df.groupby("Age Group").agg(
    total_patients=("Age Group", "count")
).reset_index()

print(query3_age_group)
```

```python
query4_insurance = df.groupby("Insurance Provider").agg(
    total_patients=("Insurance Provider", "count"),
    total_billing=("Billing Amount", "sum"),
    avg_billing=("Billing Amount", "mean")
).reset_index()

print(query4_insurance)
```

```python
query5_test_results = df.groupby("Test Results").agg(
    total_patients=("Test Results", "count"),
    avg_billing=("Billing Amount", "mean"),
    avg_stay=("Length_of_Stay", "mean")
).reset_index()

print(query5_test_results)
```

```python
with pd.ExcelWriter("healthcare_queries.xlsx") as writer:
    query1_conditions.to_excel(writer, sheet_name="query1_conditions", index=False)
    query2_admission.to_excel(writer, sheet_name="query2_admission", index=False)
    query3_age_group.to_excel(writer, sheet_name="query3_age_group", index=False)
    query4_insurance.to_excel(writer, sheet_name="query4_insurance", index=False)
    query5_test_results.to_excel(writer, sheet_name="query5_test_results", index=False)

print("File saved successfully!")
```
